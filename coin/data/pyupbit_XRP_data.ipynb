```python
#OHLCV(시가, 고가, 저가, 종가, 거래량)
df = pyupbit.get_ohlcv("KRW-XRP")
print(df.tail()) # 데이터프레임의 마지막 5행을 출력하여 가장 최근의 데이터 상태를 확인
```

```python
df = pyupbit.get_ohlcv(ticker="KRW-XRP", interval='minute10', count=52567, to="20240729", period=1)
```
